# Word Cloud

```python
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
with open("data/A.txt", 'r') as f: filedata = f.read()
```

```python
stopwords = set(STOPWORDS)
wordcloud = WordCloud(stopwords=stopwords, max_words=25, \
            background_color="white").generate(filedata)

plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```

```python
stopwords.update(["many", "using", "want", "value", "become"])
wordcloud = WordCloud(stopwords=stopwords, max_words=15, \
            background_color="black").generate(filedata)

plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```
